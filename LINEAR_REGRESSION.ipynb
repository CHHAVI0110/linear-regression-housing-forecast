## PREDICTIVE_HOUSING_ANALYSIS

```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

PROCESS_1 : LOAD THE DATASET




```python
from sklearn.datasets  import fetch_california_housing
```

```python
df=fetch_california_housing()
```

```python
df
```

```python
dataset = pd.DataFrame(df.data)
```

```python
dataset
```

```python
dataset.columns=df.feature_names
```

```python
dataset.head()
```

PROCESS_2 : EXTRACT INPUT AND OUTPUT

```python
x=dataset
```

```python
x
```

```python
y=df.target
y
```

PROCESS_3 : TRAIN A LINEAR REGRESSION MODEL

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train
```

```python
x_test
```

PROCESS_4 : SCALE THE VALUE

---



```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
x_train=scaler.fit_transform(x_train)
x_train
```

```python
x_test=scaler.transform(x_test)
x_test
```

PROCESS_5 : MODEL TRAINING USING THE LINEAR REGRESSION

```python
from sklearn.linear_model import LinearRegression
# cross validation
from sklearn.model_selection import cross_val_score
```

```python
regression=LinearRegression()
```

```python
mse=cross_val_score(regression,x_train,y_train,scoring='neg_mean_squared_error',cv=4)
mse
```

```python
np.mean(mse)
```

```python
regression.fit(x_train,y_train)
```

PROCESS_6 : PREDICTION

```python
prediction=regression.predict(x_test)
prediction
```

PROCESS_7 : ACCURACY CHECK

```python
import seaborn as snc
```

```python
snc.displot(prediction-y_test,kind='kde')
```

```python
from sklearn.metrics import r2_score
scorer=r2_score(y_test,prediction)
scorer
```

**R-squared score indicates model accuracy.**
